## Creation de Mapping pour les id

```python
# comme tout les utilisateurs n'ont pas noté tout les jobs et en plus on a filtré les données,
# les id ne sont pas continus, mais keras fonctionne avec des id continus

def generate_id_mappings(ids_list):
    userId_map = {new_id: old_id for new_id, old_id in enumerate(ids_list)}
    inverse_userId_map = {old_id: new_id for new_id, old_id in enumerate(ids_list)}
    return userId_map, inverse_userId_map
```

```python
# utilisation de notre fonction de mapping des id

userId_map, inverse_userId_map = generate_id_mappings(ratings.employe_id.unique())
jobId_map, inverse_jobId_map = generate_id_mappings(ratings.offre_id.unique())

ratings['mEmploye_id'] = ratings['employe_id'].map(inverse_userId_map)
ratings['mOffre_id'] = ratings['offre_id'].map(inverse_jobId_map)
```

## Recommandation basé sur le contenu

```python
# creation d'une matrice contentant les attributs des offres et les competences (listées une à une) en colonnes
jobs_skills_df = offres_df.join(offres_df['competencesRequises'].str.get_dummies(','))
jobs_skills_df['mOffre_id'] = jobs_skills_df['id'].map(inverse_jobId_map)

# definition des mOffre_id comme les indexes
jobs_skills_df.set_index('mOffre_id', inplace=True)

# suppression des lignes avec valeurs manquantes
jobs_skills_df = jobs_skills_df.loc[jobs_skills_df.index.dropna()]

jobs_skills_df
```

### Calcul de la similarité entre les jobs

```python
## similarité basé sur les competences requises de chaque job

from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer()
count_matrix = count.fit_transform(offres_df['competencesRequises'])
cosine_sim = cosine_similarity(count_matrix)
print(cosine_sim)
```

```python
## Recherche de l'index pour un job données
def get_index_from_id(id):
    return offres_df[offres_df["id"] == id].index.values[0]

test_jobId = 1
id = get_index_from_id(test_jobId)
```

```python
## Creation d'une liste de similarité entre le job données et les autres jobs (basé sur les competences requises)
similar_jobs = list(enumerate(cosine_sim[id]))

## Filtrage de la liste dans l'ordre decroissant
## Pour mettre les jobs les plus similaires au debut de la liste
sorted_similar_jobs = sorted(similar_jobs, key=lambda x:x[1], reverse=True)
if (id == sorted_similar_jobs[0][0]):
    sorted_similar_jobs.pop(0)

sorted_similar_jobs
```

```python
## Definition d'un seuil permettant de filtrer les jobs similaires et de recuperer ceux superieur au seuil
seuil = 0.5
def filter(seuil, similarJobs) :
    jobs_similarity = [simValue[1] for simValue in similarJobs]
    for i in range(len(jobs_similarity)):
        if (jobs_similarity[i]<= seuil):
            similarJobs = similarJobs[:i-1]
        i = len(jobs_similarity)
    return similarJobs

sorted_similar_jobs = filter(seuil, sorted_similar_jobs)
sorted_similar_jobs
```

```python
## Recuperation des jobs à recommander
topJobsId = [item[0] for item in sorted_similar_jobs]
recommended_jobs = offres_df.query('index in @topJobsId')

recommended_jobs
```

```python
## Affichage de l'index et de la valeur de similarite des jobs à recommander par rapport au job d'entrée
df = pd.DataFrame(sorted_similar_jobs[1:], columns=["index", "similarity"])
df.describe()
```

```python
## SVD
from scipy.sparse import csr_matrix
from scipy.linalg import sqrtm

R_df = ratings.pivot(index='mEmploye_id', columns="mOffre_id", values="rating")
R_df.head()

## Remplacement des valeurs manquantes par -1 (pour dire qu'il n'y a pas encore de note)
```

```python
## Etant donnée qu'on peut avoir des utilisateurs qui soit notent très bien tout les films, soit les notes très mal;
## On va alors normaliser les lignes par l'utilisateur (enlever le biais de tout les ustilisateurs)
## Ce qui signifie elever la moyen de toutes les notes des utilisateurs pour la phase d'apprentissage
## Ensuite on les ajoutera aux predictions
R = R_df.values
user_ratings_mean = np.mean(R, axis=1)
R_demeanded = R - user_ratings_mean.reshape(-1, 1)
print(R_demeanded)
```

```python
latent_dimension = 2
U, sigma, Vt = svds(R_demeanded, k = latent_dimension)

print("Dimension de U: %d * %d" % (U.shape))
```
